```python
import json

fd = open("products.json",'r')
r = fd.read()
fd.close()

products = json.loads(r)
```

```python
products
```

```python
prod  = str(input("Enter the product Id: "))
quant = int(input("Enter the product quantity: "))

print("Product: ", products[prod]['name'])
print("Price: ", products[prod]['price'])
print("Brand:", products[prod]['brand'])
print("Total Amount: ", products[prod]['price'] * quant)

products[prod]['quantity'] = products[prod]['quantity'] - quant
```

```python
pro = json.dumps(products)

fd = open("products.json",'w')
fd.write(pro)
fd.close()
```

```python
prodid = str(input("Enter product id:"))
name = str(input("Enter name:"))
price = int(input("Enter price:"))
brand = str(input("Enter brand:"))
quantity = int(input("Enter product quantity:"))

products[prodid] = {'name': name, 'price': price, 'quantity': quantity, 'brand': brand}

pro = json.dumps(products)

fd = open("products.json",'w')
fd.write(pro)
fd.close()
```
